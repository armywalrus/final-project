### Pollution Data

```python
# Read our data into pandas

pollution_2011 = pd.read_csv(file12)
pollution_2012 = pd.read_csv(file13)
pollution_2013 = pd.read_csv(file14)
pollution_2014 = pd.read_csv(file15)
pollution_2015 = pd.read_csv(file16)
```

```python
pollution_2011.reset_index(drop=True)
```

```python
# Sort Pollution dataframes by state name

pollution_2011 = pollution_2011.sort_values("State")  
pollution_2012 = pollution_2012.sort_values("State")  
pollution_2013 = pollution_2013.sort_values("State")  
pollution_2014 = pollution_2014.sort_values("State")  
pollution_2015 = pollution_2015.sort_values("State")  
```

```python
# Idenitfy rows with Mexico data

pollution_2011.iloc[150:160]
```

```python
# Get names of indexes for which column State has value of Mexico
indexNames = pollution_2011[pollution_2011['State'] == "Country Of Mexico" ].index
indexNames2 = pollution_2012[pollution_2012['State'] == "Country Of Mexico" ].index
indexNames3 = pollution_2013[pollution_2013['State'] == "Country Of Mexico" ].index
indexNames4 = pollution_2014[pollution_2014['State'] == "Country Of Mexico" ].index
indexNames5 = pollution_2015[pollution_2015['State'] == "Country Of Mexico" ].index


# Delete these row indexes from dataFrame
pollution_2011.drop(indexNames , inplace=True)
pollution_2012.drop(indexNames2 , inplace=True)
pollution_2013.drop(indexNames3 , inplace=True)
pollution_2014.drop(indexNames4 , inplace=True)
pollution_2015.drop(indexNames5 , inplace=True)

pollution_2011.iloc[150:160]
```

```python
# Sort Pollution dataframes by state name

pollution_2011_states = pollution_2011.sort_values("State")  
pollution_2012_states = pollution_2012.sort_values("State")  
pollution_2013_states = pollution_2013.sort_values("State")  
pollution_2014_states = pollution_2014.sort_values("State")  
pollution_2015_states = pollution_2015.sort_values("State")  
```

```python
# Append latitude and longitude into Pollution dataframes

pollution_2011_states['Latitude'] = states['latitude']
pollution_2011_states['Longitude'] = states['longitude']
pollution_2012_states['Latitude'] = states['latitude']
pollution_2012_states['Longitude'] = states['longitude']
pollution_2013_states['Latitude'] = states['latitude']
pollution_2013_states['Longitude'] = states['longitude']
pollution_2014_states['Latitude'] = states['latitude']
pollution_2014_states['Longitude'] = states['longitude']
pollution_2015_states['Latitude'] = states['latitude']
pollution_2015_states['Longitude'] = states['longitude']
```

```python
# Stack tables together using concat

pollution_concat_1 = pd.concat([pollution_2011_states, pollution_2012_states], axis =0)
pollution_concat_2 = pd.concat([pollution_concat_1, pollution_2013_states], axis =0)
pollution_concat_3 = pd.concat([pollution_concat_2, pollution_2014_states], axis =0)
pollution_all = pd.concat([pollution_concat_3, pollution_2015_states], axis =0)
pollution_all.iloc[150:160]
```

```python
# Get list of column names

list(pollution_all.columns) 
```

```python
# Group Data by State

pollution_stats = pd.DataFrame({"Days_with_AQI": pollution_all.groupby(["State"]).mean()["Days_with_AQI"], 
                                "Year": pollution_all.groupby(["State"]).count()["Year"],
                             "Good_Days": pollution_all.groupby(["State"]).mean()["Good_Days"],
                             "Moderate_Days": pollution_all.groupby(["State"]).mean()["Moderate_Days"], 
                             "Unhealthy_Days": pollution_all.groupby(["State"]).mean()["Unhealthy_Days"],   
                             "Very_Unhealthy_Days": pollution_all.groupby(["State"]).mean()["Very_Unhealthy_Days"], 
                             "Hazardous_Days": pollution_all.groupby(["State"]).mean()["Unhealthy_Days"],
                             "Days_CO": pollution_all.groupby(["State"]).mean()["Days_CO"],
                             "Days_NO2": pollution_all.groupby(["State"]).mean()["Days_NO2"],
                             "Days_Ozone": pollution_all.groupby(["State"]).mean()["Days_Ozone"],
                             "Days_SO2": pollution_all.groupby(["State"]).mean()["Days_SO2"],
                             "Days_PM2.5": pollution_all.groupby(["State"]).mean()["Days_PM2.5"],
                             "Days_PM10": pollution_all.groupby(["State"]).mean()["Days_PM10"],
                             "Latitude": pollution_all.groupby(["State"]).mean()["Latitude"],
                             "Longitude": pollution_all.groupby(["State"]).mean()["Longitude"]})                                                                            
                                                                                     
                                                                                    
pollution_stats.head()
```

```python
# Export to csv

pollution_stats.reset_index("State").to_csv('..\Resources\Tableau_clean\pollution_stats.csv', index=False)
```
